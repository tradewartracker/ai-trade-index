```python
# Cost estimation for classifying 22,645 HS10 codes

# Claude Sonnet 4 pricing (as of Jan 2026)
INPUT_COST_PER_MILLION = 3.00   # $3 per million input tokens
OUTPUT_COST_PER_MILLION = 15.00  # $15 per million output tokens

# Token estimates per classification request
SYSTEM_PROMPT_TOKENS = 350      # The system prompt
TOOL_DEFINITION_TOKENS = 200    # The classification tool schema
USER_MESSAGE_BASE = 20          # "Classify this HS10 code:\n\nCode: ... Description: ..."
AVG_DESCRIPTION_TOKENS = 80     # Average HS10 description length

# Input tokens per request
INPUT_TOKENS_PER_REQUEST = SYSTEM_PROMPT_TOKENS + TOOL_DEFINITION_TOKENS + USER_MESSAGE_BASE + AVG_DESCRIPTION_TOKENS

# Output tokens (max_tokens=400, but actual usage varies)
AVG_OUTPUT_TOKENS_PER_REQUEST = 250  # Conservative estimate

# Total calculations
TOTAL_CODES = 22645

total_input_tokens = TOTAL_CODES * INPUT_TOKENS_PER_REQUEST
total_output_tokens = TOTAL_CODES * AVG_OUTPUT_TOKENS_PER_REQUEST

input_cost = (total_input_tokens / 1_000_000) * INPUT_COST_PER_MILLION
output_cost = (total_output_tokens / 1_000_000) * OUTPUT_COST_PER_MILLION
total_cost = input_cost + output_cost

print("=" * 60)
print("COST ESTIMATION FOR CLASSIFYING 22,645 HS10 CODES")
print("=" * 60)
print(f"\nModel: Claude Sonnet 4 (claude-sonnet-4-20250514)")
print(f"\nTotal codes to classify: {TOTAL_CODES:,}")
print(f"\nPer-request estimates:")
print(f"  - Input tokens:  ~{INPUT_TOKENS_PER_REQUEST} tokens")
print(f"  - Output tokens: ~{AVG_OUTPUT_TOKENS_PER_REQUEST} tokens")
print(f"\nTotal tokens:")
print(f"  - Input:  {total_input_tokens:,} tokens ({total_input_tokens/1_000_000:.2f}M)")
print(f"  - Output: {total_output_tokens:,} tokens ({total_output_tokens/1_000_000:.2f}M)")
print(f"\nCost breakdown:")
print(f"  - Input cost:  ${input_cost:,.2f}")
print(f"  - Output cost: ${output_cost:,.2f}")
print(f"  - TOTAL COST:  ${total_cost:,.2f}")
print(f"\nEstimated time (with 0.5s delay): {(TOTAL_CODES * 0.5 / 3600):.1f} hours")
print("=" * 60)
```
