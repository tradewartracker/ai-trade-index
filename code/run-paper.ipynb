# Run Paper Pipeline (Notebooks 06–11)
Executes all paper-generating notebooks in order.  
**Notebook 06 initializes** `ai-trade-results.tex`; **07–11 append** to it.

Run the cell below to regenerate all tables, figures, and LaTeX macros.

```python
import subprocess, sys, time

notebooks = [
    '06-section2-products.ipynb',
    '07-section2-growth.ipynb',
    '08-section2-countries.ipynb',
    '09-section2-tariffs.ipynb',
    '10-section3-exports.ipynb',
    '11-counterfactual.ipynb',
]

for nb in notebooks:
    print(f'\n{"="*60}')
    print(f'Running {nb}...')
    print(f'{"="*60}')
    t0 = time.time()
    result = subprocess.run(
        [sys.executable, '-m', 'jupyter', 'nbconvert',
         '--to', 'notebook', '--execute',
         '--ExecutePreprocessor.timeout=600',
         '--inplace', nb],
        capture_output=True, text=True
    )
    elapsed = time.time() - t0
    if result.returncode == 0:
        print(f'  OK  ({elapsed:.1f}s)')
    else:
        print(f'  FAILED  ({elapsed:.1f}s)')
        print(result.stderr[-2000:])
        break

print(f'\n{"="*60}')
print('Done.' if result.returncode == 0 else 'Stopped on error.')
print(f'{"="*60}')
```
